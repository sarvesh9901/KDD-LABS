```python
import pandas as pd

path = r"E:\DYPIU\SEM 6\KDD\KDD LABS\LAB_03\kc_house_data.csv"
missing_val_format = ["n.a","?","NA","n/a","na","--","NaN"]
df = pd.read_csv(path,na_values=missing_val_format)
df.head()
```

# Normalization 

```python
from numpy import asarray
from sklearn.preprocessing import MinMaxScaler

#data as array.
data  = df[["price","bedrooms","bathrooms","sqft_living","floors"]].to_numpy()
data
#define scaler

scaler = MinMaxScaler()
scaler = scaler.fit(data)

scaler.data_min_
scaler.data_max_

scaled = scaler.transform(data)
df2 = pd.DataFrame(scaled,columns=["price","bedrooms","bathrooms","sqft_living","floors"])
df2.head()
```

# Standardization 

```python
from numpy import asarray
from sklearn.preprocessing import StandardScaler

#data as array.
data  = df[["price","bedrooms","bathrooms","sqft_living","floors"]].to_numpy()
data
#define scaler

scaler = StandardScaler()
scaler = scaler.fit(data)

scaler.mean_
scaler.scale_

scaled = scaler.transform(data)
df2 = pd.DataFrame(scaled,columns=["price","bedrooms","bathrooms","sqft_living","floors"])
df2.head()
```
